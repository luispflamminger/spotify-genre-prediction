```python
import json
import pandas as pd
import os

#Loading data
f = open(os.path.join(os.getcwd(), "json", "flat_json_no_features.json"))
data = json.load(f)

df = pd.json_normalize(data)
df
```

```python
df.groupby(by='category.id').agg('count').sort_values("category.name", ascending=False)
```

```python
df_no_duplicates = df.drop_duplicates(subset=['category.playlist.track.id'], keep='last')
df_no_duplicates.groupby(by='category.id').agg('count').sort_values("category.name", ascending=False)
```
