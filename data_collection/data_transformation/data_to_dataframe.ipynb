```python
import json
import pandas as pd
import os

#Loading data
f = open(os.path.join(os.getcwd(), "..", "json", "playlists.json"))
data = json.load(f)

#Experimenting with datatypes
print(type(data) == dict)
print(type(data["categories"]) == list)
print(type(data["categories"][0]["playlists"]))
print(type(data["categories"][0]["playlists"][0]["id"]))
#dict -> list -> dict -> list -> dict -> string

df_normalized = pd.json_normalize(data["categories"])
df_normalized
#only normalizes to first level
```

```python
#recursive function to flatten json
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '.')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '.')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out

flat_json = flatten_json(data["categories"])
#with open(os.path.join(os.getcwd(), "json", "flat.json"), 'w') as outfile:
#    json.dump(data, outfile, indent=2)
df_flattened = pd.json_normalize(flat_json)
df_flattened
#format not usable like this
```

```python

# Solution using code from stackoverflow
# https://stackoverflow.com/questions/52795561/flattening-nested-json-in-pandas-data-frame
# Currently doesn't support multiple nested dicts

# normalize events
df_1 = pd.json_normalize(data, 'categories')

# explode all columns with lists of dicts
df_1 = df_1.apply(lambda x: x.explode()).reset_index(drop=True)

# list of columns with dicts
cols_to_normalize = ['playlists']

# if there are keys, which will become column names, overlap with excising column names
# add the current column name as a prefix
normalized = list()
for col in cols_to_normalize:
    
    d = pd.json_normalize(df_1[col], sep='_')
    d.columns = [f'{col}_{v}' for v in d.columns]
    normalized.append(d.copy())

# combine df with the normalized columns
df_1 = pd.concat([df_1] + normalized, axis=1).drop(columns=cols_to_normalize)
df_1
```
